```python
import pandas as pd

# Initialize a list to hold our pathway data
results = []

# Loop through each pathway in the list stored at df['pathways'][0]
for pathway in df['pathways'][0]:
    # Extract the pathway name (use a default if not available)
    pathway_name = pathway.get('name', 'Unnamed Pathway')
    reactome_id = pathway.get('stId', 'Unknown ID')

    # Extract gene IDs from the entities list; each entity is expected to have an 'id'
    gene_ids = [entity.get('id') for entity in pathway.get('data', {}).get('entities', [])]

    results.append({
        'pathway': pathway_name,
        'reactome_id': reactome_id,
        'genes': gene_ids
    })

pathway_df = pd.DataFrame(results)

pathway_df.to_csv(f"GenesInPathwaysFor{file_path.split('.')[0]}.csv", index=False)
```
